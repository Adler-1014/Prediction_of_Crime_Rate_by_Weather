```python
import csv
import math
```

```python
f = open('서울데이터 - 복사본.csv', 'r', encoding='utf-8')
data = csv.reader(f)
next(data)
data = list(data)
```

```python
f.close()
```

```python
for i in range(len(data)):
    for j in range(len(data[i])):
        if data[i][j] == '':
            data[i][j] = '0'
        data[i][j] = float(data[i][j].replace(',',''))
```

```python
for i in data:
    for j in i:
        print(j, type(j))
```

```python
import torch
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler

torch.manual_seed(0)
```

```python
t = int(len(data) * 0.3)
t = int(len(data) - t)
t
```

```python
xdata = data[:t]
ydata = data[t:]

#for i in xdata:
#    print(i)
```

```python
data_dim = 6
hidden_dim = 100
output_dim = 7
learning_rate = 0.001
iterations = 500
```

```python
class Net(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, layers):
        super(Net, self).__init__()
        self.rnn = torch.nn.LSTM(input_dim, hidden_dim, num_layers=layers, batch_first=True)
        self.fc = torch.nn.Linear(hidden_dim, output_dim, bias=True)
    
    def forward(self, x):
        x, _status = self.rnn(x)
        x = self.fc(x[:, -1])
        return x
    
net = Net(data_dim, hidden_dim, output_dim, 1)

criterion = torch.nn.MSELoss()
optimizer = optim.Adam(net.parameters(), lr=learning_rate)
```

```python
trainX = []
trainY = []
for i in xdata:
    trainY.append([i[1:8]])
    trainX.append([i[8:]])
```

```python
for i in range(len(trainY)):
    for j in range(len(trainY[i][0])):
        try:
            trainY[i][0][j] =  math.log10(float(trainY[i][0][j]))
        except:
            trainY[i][0][j] =  float(trainY[i][0][j])
            
for i in range(len(trainX)):
    for j in range(len(trainX[i][0])):
        trainX[i][0][j] = float(trainX[i][0][j])
```

```python
testX = []
testY = []
for i in ydata:
    testY.append([i[1:8]])
    testX.append([i[8:]])

plt.plot([i for i in range(len(trainY[0][0]))], trainY[0][0], [i for i in range(len(testY[0][0]))], testY[0][0])
plt.show()
    
for i in range(len(testY)):
    for j in range(len(testY[i][0])):
        try:
            testY[i][0][j] =  math.log10(float(testY[i][0][j]))
            print(testY[i][0][j])
        except:
            testY[i][0][j] =  float(testY[i][0][j])
            
for i in range(len(testX)):
    for j in range(len(testX[i][0])):
        testX[i][0][j] = float(testX[i][0][j])
```

```python
trainX_tensor = torch.FloatTensor(trainX)
trainY_tensor = torch.FloatTensor(trainY)

testX_tensor = torch.FloatTensor(testX)
testY_tensor = torch.FloatTensor(testY)

#testX_tensor = torch.FloatTensor(testX)
#testY_tensor = torch.FloatTensor(testY)
```

```python
plt.plot([i for i in range(len(trainY[0][0]))], trainY[0][0], [i for i in range(len(testY[0][0]))], testY[0][0])
plt.show()
```

```python
losses = []
for i in range(iterations):

    optimizer.zero_grad()
    
    outputs = net(trainX_tensor)
    loss = criterion(outputs, trainY_tensor)
    
    loss.backward()
    optimizer.step()
    print(i, loss.item())
    losses.append(loss.item())
```

```python
from matplotlib import pyplot as plt
 
plt.plot([i for i in range(500)], losses)
plt.show()
```
